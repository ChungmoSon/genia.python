```python
s=metro['역명']=='서울역'
y=metro['역명']=='역삼'

s_mean_in=metro[s][metro['승하차구분']=='승차']['승하차인원수'].mean()
y_mean_in=metro[y][metro['승하차구분']=='승차']['승하차인원수'].mean()
s_mean_out=metro[s][metro['승하차구분']=='하차']['승하차인원수'].mean()
y_mean_out=metro[y][metro['승하차구분']=='하차']['승하차인원수'].mean()

metro_in=(s_mean_in+y_mean_in)/2
metro_out=(s_mean_out+y_mean_out)/2

plt.bar(['승차','하차'],[metro_in,metro_out])
plt.ylim(95000,97000)
plt.text(-0.08,metro_in+100,round(metro_in))
plt.text(0.88,metro_in+100,'%.2f' %metro_out)
```
