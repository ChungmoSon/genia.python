```python
from scipy import stats

for x in df.index:
  if df.loc[x, "일일대당원가"] < 10000:
    df.drop(x, inplace = True)
```

```python
bus1 = df[df['업체아이디']==1000]['일일대당원가']
```

```python
bus2 = df[df['업체아이디']==1047]['일일대당원가']
```

```python
stats.ttest_ind(bus1, bus2)
```

```python
stats.ks_2samp(bus1, bus2)
```
