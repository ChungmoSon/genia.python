## 실습 - 그래프 다루기
* 서울역, 역삼역에서 승차하는 인원과 하차하는 평균 인원 숫자를 그래프로 나타내기
* 단, 각 그래프의 꼭대기 부분에 승하차인원의 평균값을 명시한다

```python
s = df['역명'] == '서울역'
y = df['역명'] == '역삼'
```

```python
s_mean = df[s][df['승하차구분'] == '승차']['승하차인원수'].mean()
y_mean = df[y][df['승하차구분'] == '승차']['승하차인원수'].mean()
metro_in = (s_mean + y_mean) / 2
```

```python
metro_out = (df[s][df['승하차구분'] == '하차']['승하차인원수'].mean() + df[y][df['승하차구분'] == '하차']['승하차인원수'].mean()) / 2
```

```python
plt.bar(['승차','하차'], [metro_in, metro_out])
plt.ylim(95000,97000)
plt.text(-0.08, metro_in+100, round(metro_in))
plt.text(0.88, metro_out+100, '%.2f' %metro_out)
```
