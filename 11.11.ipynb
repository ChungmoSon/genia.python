```python
import pandas as pd
import numpy as np
df=pd.read_csv('test.csv')
df
```

```python
df['check column'] = ''
df['Sex'] = ''
mean_sum1=0
mean_count1=0
mean_sum2=0
mean_count2=0
fare=0
count1=0
count2=0
count3=0
male_sum=0

for i in df.index:
    
    if ('Mrs' or 'Miss' or 'Ms') in df['Name'][i]:
        df['Sex'][i] = 'female'
    elif 'Mr' in df['Name'][i]:
        df['Sex'][i] = 'male'#a
    else:
        fare+=df['Fare'][i]
    
    if pd.isnull(df['Age'][i]):
        df['Age'][i] = 0 #b
    
    if df['Sex'][i] == 'male' and 10 <= df['Age'][i] and df['Age'][i] < 20:
        df['check column'][i] = 'O'
        mean_sum2 += df['Age'][i]
        mean_count2 += 1
    else:
        df['check column'][i] = 'X' #c
        
    if df['Sex'][i] == 'female' and df['Age'][i] >= 20:
        mean_sum1 += df['Age'][i]
        mean_count1 += 1
        
    if (('Master' or 'Dr' or 'Pf' or 'MD') in df['Name'][i]) and df['Pclass'][i] == 1:
        count1 +=1
    elif (('Master' or 'Dr' or 'Pf' or 'MD') in df['Name'][i]) and df['Pclass'][i] ==2:
        count2 +=1
    elif (('Master' or 'Dr' or 'Pf' or 'MD') in df['Name'][i]) and df['Pclass'][i] ==3:
        count3 +=1
    
    if df['Sex'][i] == 'male':
        male_sum += df['Fare'][i]
         
print("The percentage of the fare for people of unknown sex is", fare/df['Fare'].sum()*100,"%")
        
if mean_sum1/mean_count1 > mean_sum2/mean_count2:
    print('Number of female and over 20(',mean_sum1/mean_count1,') is bigger than male and teenager(',mean_sum2/mean_count2,')')
else: 
    print('Number of female and over 20(',mean_sum1/mean_count1,') is smaller than male and teenager(',mean_sum2/mean_count2,')')
        
df.to_csv('result.tsv', sep ='\t', index=False) #d

print("Number of people who own over a master degree in class 1 is",count1)
print("Number of people who own over a master degree in class 2 is",count2)
print("Number of people who own over a master degree in class 3 is",count3)

df[5:15]
```
