```python
def check_spl(x):
    if x[0].isalpha() or x[0].isdigit() :
        return False 
    else:
        return True
    
def check_eng(x):
    if x[0].isascii() and x[0].isalpha() :
        return True
    else:
        return False

def check_kor(x):
    if not x[0].isascii() and x[0].isalpha():
        return True
    else:
        return False
    
m1=df['가맹점명'].apply(lambda x:x[0].isdigit())#number
m2=df['가맹점명'].apply(lambda x:check_eng(x))#English
m3=df['가맹점명'].apply(lambda x:check_spl(x))#symbols
m4=df['가맹점명'].apply(lambda x:check_kor(x))#Korean
df_num=df[m1].sort_values('가맹점명')
df_eng=df[m2].sort_values('가맹점명')
df_spl=df[m3].sort_values('가맹점명')
df_kor=df[m4].sort_values('가맹점명')
df_res=pd.concat([df_kor,df_eng,df_num,df_spl])
```
